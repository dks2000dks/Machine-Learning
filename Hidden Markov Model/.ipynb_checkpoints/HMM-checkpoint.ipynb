## Pre-Processing Data

Finding Mel-Frequency Cepstral Coefficients (MFCCs) from the raw speech samples. Picking 25 ms worth of speech samples with a 10 ms overlap to find MFCCs. Using the basic 13 element version of MFCC as the feature vector representing 25 ms of speech.

### Selecting Phones

```python
PhonesPath = "Data/SVL-PRASANNA-CV/"
PhonesAvailable = os.listdir(PhonesPath)
print ("Available Phones:\n", PhonesAvailable)
PhonesSelected = ["ga", "te"]
print ("------------------------------------------")
print ("Selected Phones:\n", PhonesSelected)
```

### Reading Data

```python
def ReadData(Path,SplitRatio=0.1):
    Files = sorted(os.listdir(Path))
    Data = []
    for f in Files:
        Rate,Signal = wav.read(Path + "/" + f)
        D = mfcc(signal = Signal,
                    winlen = 0.025,
                    winstep = 0.015) # As we need a overlap of 10ms
        D = (D-np.mean(D))/(np.max(D) - np.min(D))
        Data.append(D)
    Data = np.asarray(Data)
    
    m = int(Data.shape[0] * SplitRatio)
    
    TestData = Data[:m]
    TrainData = Data[m:]
    
    return TrainData,TestData

TrainPhone0,TestPhone0 = ReadData(PhonesPath + PhonesSelected[0])
TrainPhone1,TestPhone1 = ReadData(PhonesPath + PhonesSelected[1])
```

## Hidden Markov Model

Building a **Hidden Markov Model with Forward Backward** (also known as the Baum-Welch) algorithm to learn the Parameters of a HMM.

**Parameters:**
\begin{align}
A_{jk} = p(z_{nk} = 1 | z_{(n-1)j} = 1) \\
\pi_{k} = p(z_{1k}) \\
p(x_{n}|z_{n} = k, \phi) = p(x_{n}|\phi_{k})
\end{align}

**Likelihood Function:**
\begin{align}
Q(\theta, \theta^{old}) = \sum_{k=1}^{K} \gamma\left(z_{1 k}\right) \ln \pi_{k}+\sum_{n=2}^{N} \sum_{j=1}^{K} \sum_{k=1}^{K} \xi\left(z_{n-1, j}, z_{n k}\right) \ln A_{j k} +\sum_{n=1}^{N} \sum_{k=1}^{K} \gamma\left(z_{n k}\right) \ln p\left(x_{n} \mid \phi_{k}\right)
\end{align}
\
For simplicity, we consider $\log p(X)$ as LogLikelihood Function.

**Forward-Backward Algorithm:**
\begin{align}
c_{n} \widehat{\alpha}\left(z_{n}\right)=p\left(x_{n} \mid z_{n}\right) \sum_{z_{n-1}} \widehat{\alpha}\left(z_{n-1}\right) p\left(z_{n} \mid z_{n-1}\right) \\
c_{n+1} \widehat{\beta}\left(z_{n}\right)=\sum_{z_{n+1}} \widehat{\beta}\left(z_{n+1}\right) p\left(x_{n+1} \mid z_{n+1}\right) p\left(z_{n+1} \mid z_{n}\right)
\end{align}

**Posterior Probabilities:**
\begin{align}
\gamma(z_{n}) = \widehat{\alpha}(z_{n})\widehat{\beta}(z_{n})\\
\xi\left(z_{n-1}, z_{n}\right)=c_{n} \widehat{\alpha}\left(z_{n-1}\right) p\left(x_{n} \mid z_{n}\right) p\left(z_{n} \mid z_{n-1}\right) \hat{\beta}\left(z_{n}\right) \\
p(X) = \prod_{n=1}^{N} c_{n}
\end{align}

**Estimating Parameters:**
\begin{align}
\mu_{k}=\frac{\sum_{n=1}^{n} \gamma\left(z_{n k}\right) x_{n}}{\sum_{n=1}^{n} \gamma\left(z_{n k}\right)}=\frac{\sum_{n=1}^{n} \alpha\left(z_{n k}\right) \beta\left(z_{n k}\right) x_{n}}{\sum_{n=1}^{n} \alpha\left(z_{n k}\right) \beta\left(z_{n k}\right)} \\
\Sigma_{k}=\frac{\sum_{n=1}^{N} \gamma\left(z_{n k}\right)\left(x_{n}-\mu_{k}\right)\left(x_{n}-\mu_{k}\right)^{\mathrm{T}}}{\sum_{n=1}^{N} \gamma\left(z_{n k}\right)} \\
\pi_{k} = \frac{\gamma\left(z_{1 k}\right)}{\sum_{j=1}^{K} \gamma\left(z_{1 j}\right)} \\
A_{j k} = \frac{\sum_{n=2}^{N} \xi\left(z_{n-1, j}, z_{n k}\right)}{\sum_{l=1}^{K} \sum_{n=2}^{N} \xi\left(z_{n-1, j}, z_{n l}\right)}
\end{align}

**Initialisations:**
\begin{align}
\widehat{\alpha}(z_{1k}) = \pi_{k}p(x_{1}|z_{1k}) \\
\widehat{\beta}(z_{Nk}) = 1
\end{align}

**Procedure:**
1. Initialise Parameters $\{\pi, A, \phi\}$.
2. **E-Step**: Calculate LogLikelihood
3. Calculate $\alpha$ and $\beta$ and then $\gamma$ and $\xi$.
4. **M-Step**: Estimate $\pi_k$, $A$, $\mu_{k}$ and $\sum_{k}$.


---

**Parameter Descriptions:**
- K: No.of **States of Latent Variables**.
- X: **MFCC Coefficients** of All Audio Smaples.
- N: No.of **Time-Steps.**
- M: No.of Data Points.
- $A_{jk}$: **Transition Probability** or Conditional Probability for Transition among Latent Variables.
- $\pi_{k}$: Probabailites of first Latent Variable $z_{1}$.
- $\phi = \{\mu, \sum\}$: State Dependent Parametrisation and $\phi$ is considered to be Gaussian Distribution.
- $p(x_{n}|z_{n} = k, \phi)$: **Emission Probabilities**.
- $Q(\theta, \theta^{old})$: **Likelihood Function**.
- $\alpha, \beta$: Terms of Forward-Backward Algorithm.
- $\gamma$: Marginal Posterior Distribution of Latent Variable $z_{n}$.
- $\xi$: Joint Posterior Distribution of two successive Latent Variables $z_{n}, z_{n-1}$.

```python
class HMM():
    def __init__(self,K):
        self.K = K
        self.N = None
        self.D = None
        self.M = None
        
        self.Pi = None
        self.A = None
        self.Mean = None
        self.CoVariance = None
        
        self.EmPro = None
        self.Alpha_Hat = None
        self.Beta_Hat = None
        
        self.Gamma = None
        self.Xi = None
  
    def InitialiseParameters(self):
        # Initialising Parameters
        self.Pi = np.ones((self.K,))/self.K
        
        self.A = np.zeros((self.K,self.K))
        for s in range(self.K):
            self.A[s] = np.random.uniform(0,1,self.K)
            self.A[s] = self.A[s]/np.sum(self.A[s])
            
        
        self.Mean = np.random.rand(self.K,self.D)
        for s in range(self.K):
            self.Mean[s] = self.X[np.random.randint(0,self.M-1)][np.random.randint(0,10)]
        
        self.CoVariance = np.zeros((self.K,self.D,self.D))
        for s in range(self.K):
            C = np.random.randn(self.D,self.D)
            self.CoVariance[s] = np.diag(np.diag(np.dot(C.T,C)))
        
    
    def EmissionProbabilities(self,x):
        # Calculating Emission Probabilities
        self.EmPro = np.zeros((self.K, self.N))
        for s in range(self.K):
            self.EmPro[s] = multivariate_normal.pdf(x,self.Mean[s],self.CoVariance[s],allow_singular=True)    
            
            
    def Forward(self,x):
        # Forward Propagation
        self.N = x.shape[0]
        self.EmissionProbabilities(x)
        
        self.Alpha_Hat = np.zeros((self.K,self.N))
        self.C = np.zeros((self.N))
        
        self.Alpha_Hat[:,0] = np.multiply(self.Pi, self.EmPro[:,0])
        self.C[0] = np.sum(self.Alpha_Hat[:,0])
        self.Alpha_Hat[:,0] /= self.C[0]
        
        for t in range(1,self.N):
            self.Alpha_Hat[:,t] = np.multiply(self.EmPro[:,t], np.dot(self.A.T, self.Alpha_Hat[:,t-1]))
            self.C[t] = np.sum(self.Alpha_Hat[:,t])
            self.Alpha_Hat[:,t] /= self.C[t]
            
        LogLikelihood = np.sum(np.log(self.C))
            
        return LogLikelihood
    
    
    def Backward(self):
        # Backward Propagation
        self.Beta_Hat = np.ones((self.K,self.N))
        
        for t in range(self.N-2, -1, -1):
            self.Beta_Hat[:,t] = np.dot(self.A, np.multiply(self.EmPro[:,t+1], self.Beta_Hat[:,t+1]))/self.C[t+1]
        
    
    def PosteriorProbabilities(self):
        # Calculating Posterior Probabilities
        self.Gamma = np.multiply(self.Alpha_Hat, self.Beta_Hat)
        
        self.Xi = np.zeros((self.N-1,self.K,self.K))
        for t in range(1,self.N):
            for j in range(self.K):
                for k in range(self.K):
                    self.Xi[t-1][j][k] = self.C[t] * self.Alpha_Hat[j][t-1] * self.EmPro[k][t] * self.A[j][k] * self.Beta_Hat[k][t]
        
        
    def EstimateParameters(self,x):
        # Re-Estimating Parameters
        
        # Pi
        self.Pi = self.Gamma[:,0]/np.sum(self.Gamma[:,0])
        
        # A
        self.A = np.zeros((self.K,self.K))
        for j in range(self.K):
            for k in range(self.K):
                for n in range(self.N-1):
                    self.A[j][k] += self.Xi[n][j][k]
            self.A[j] = self.A[j]/np.sum(self.A[j])
        
        # Mean
        self.Mean = np.zeros((self.K,self.D))
        for s in range(self.K):
            N,D = 0,0
            for t in range(self.N):
                N += (x[t] * self.Gamma[s][t])
                D += (self.Gamma[s][t])
            self.Mean[s] = N/D
            
        # CoVariance
        for s in range(self.K):
            N,D = 0,0
            for t in range(self.N):
                N += (self.Gamma[s][t] * np.matmul(x[t] - self.Mean[s], (x[t] - self.Mean[s]).T))
                D += (self.Gamma[s][t])
            self.CoVariance[s] = (N/D) + 0.2*np.identity(self.D)
            
        
        
    def fit(self,X,Threshold):
        # Training Model
        self.X = X
        self.M = self.X.shape[0]
        self.D = self.X[0].shape[1]
        self.LikelihoodData = []
        
        self.InitialiseParameters()
        e = 0
        
        while 1:
            e += 1
            L = []
            for x in self.X:
                self.N = x.shape[0]
                l = self.Forward(x)
                if (np.isnan(self.Mean).any() or np.isnan(self.CoVariance).any() or np.isnan(self.A).any() or np.isnan(self.Pi).any() or np.isnan(l)):
                    print ("Re-Initialising Parameters")
                    self.InitialiseParameters()
                else:
                    L.append(l)
                    self.Backward()
                    self.PosteriorProbabilities()
                    self.EstimateParameters(x)

            self.LikelihoodData.append(np.mean(L))
            print ("Epoch: %-*s  Average LogLikelihood: %-*s" % (6,e,24,np.mean(L)))
            if len(self.LikelihoodData) >= 2 and np.abs(self.LikelihoodData[-1] - self.LikelihoodData[-2]) < Threshold:
                break
            
    def PlotLogLikelihood(self):
        # Plotting LogLikelihood
        plt.figure(figsize=(10,8))
        plt.plot(self.LikelihoodData)
        plt.title("Graph of LogLikelihood vs Epochs")
        plt.xlabel("Epochs")
        plt.ylabel("Likelihood")
        plt.grid()
        plt.show()
```

## Building a Two-Class Classifier

```python
class Classifier():
    def __init__(self,K):
        self.hmm1 = HMM(K)
        self.hmm2 = HMM(K)
        
    def TrainModels(self, Data1, Data2, Threshold):
        # Training Models
        print ("Training Model-1")
        self.hmm1.fit(Data1,Threshold)
        self.hmm1.PlotLogLikelihood()
        print ()
        print ("-------------------------------------------------------------------")
        print ()
        print ("Training Model-2")
        self.hmm2.fit(Data2,Threshold)
        self.hmm2.PlotLogLikelihood()
        
    def Classify(self,X):
        # Classification
        for x in X:
            l1 = self.hmm1.Forward(x)
            l2 = self.hmm2.Forward(x)
            if l1 > l2:
                print ("Input belongs to Phone0")
            else:
                print ("Input belongs to Phone1")
```

Testing Classifier for different No.of States

```python
CF = Classifier(3)
CF.TrainModels(TrainPhone0,TrainPhone1,1e-3)

print ("-----------Testing Phone-0-----------")
CF.Classify(TestPhone0)
print()
print ("-----------Testing Phone-1-----------")
CF.Classify(TestPhone1)
```
